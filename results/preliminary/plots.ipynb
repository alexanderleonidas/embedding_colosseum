```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams["figure.dpi"] = 150
plt.rcParams["figure.figsize"] = (12, 10)
sns.set_theme("talk")
```

```python
df = pd.read_csv(r"FRQI\19-21-05\results.csv")
df["step"] = (df["epoch"] + 1) * (df["batch"] + 1)
plt.subplot(2, 1, 1)
sns.lineplot(data=df, x="step", y="train_loss", label="Train Loss", errorbar=("ci", 95))
sns.lineplot(data=df, x="step", y="val_loss", label="Validation Loss")
plt.ylim(0.3, 1.0)

plt.subplot(2, 1, 2)
sns.lineplot(
    data=df, x="step", y="train_acc", label="Train Accuracy", errorbar=("ci", 95)
)
sns.lineplot(data=df, x="step", y="val_acc", label="Validation Accuracy")
plt.ylim(0.5, 1.0)

plt.suptitle("FRQI: 50 images per batch, 3 layers, 4x4 images")
plt.tight_layout()
```
